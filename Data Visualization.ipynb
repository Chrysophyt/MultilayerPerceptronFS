```python
from data.ImageLoader import getImages, convertToGrayscale, preprocessImage
import matplotlib.pyplot as plt
```

```python
def showImages(images, label, cmap='viridis'):
    fig, axs = plt.subplots(2, 2, figsize=(8,6))

    fig.suptitle(label+" images from dataset.", fontsize=16)
    axs[0, 0].imshow(images[0], cmap=cmap)
    axs[0, 1].imshow(images[1], cmap=cmap)
    axs[1, 0].imshow(images[2], cmap=cmap)
    axs[1, 1].imshow(images[3], cmap=cmap)
```

```python
daisyImages = getImages("daisy")
print(daisyImagesGrayscale.shape)
showImages(daisyImages, "Daisy")
```

```python
daisyImagesGrayscale = convertToGrayscale(daisyImages)
print(daisyImagesGrayscale.shape)
showImages(daisyImagesGrayscale, "Daisy Grayscale", cmap='gray')
```
